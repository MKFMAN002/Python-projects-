```python
# Makofane Comfort (mkfman002)

num_words = ["","one","two","three","Four","Five","six","seven","eight","nine"]
Suffix_num = ["ten","eleven","twelve", "thirteen", "Fourteen","fifteen","sixteen","seventeen","eighteen","ninteen"]
num_with_0_words = ["", "", "twenty","thirty","forty","fifty","sixty","seventy","eighty","ninety"]

# input the number
input_num = int(input("Enter the number: ")) # input values

minus = "" # empty string to store (minus) if the number is minus
range_ = "[-9999-9999]"  

# if is negative store the string {minus} and proceed executing the proceeding blocks:
if input_num < 0:
    minus = "minus"  # Record whether the user input started with negative number or not
    input_num = abs(input_num) # IGNORE IT and deal with it later !
    
split_val = [int(j) for j in str(input_num)] # This split the values if more than one is inputted

if input_num == 0: # print zero since it not sotered in the data structure as it complicate cases such printing number like (102):
    print("Zero") #  as it will as {one hundred zero two} so i decided to leave the space instead of zero to avoid the situation
    
        
 # {UNITS CONVENVENTION}
# Thousands:
thousands = input_num//1000
dif_th = input_num - 1000*thousands

# Hundreds:
hundred = dif_th//100
diff_hund = dif_th - hundred*100

# Tens and ones:
tens = diff_hund//10
ones = diff_hund - tens*10

    #instruct the user the number inputted is not of the write size:
if len(split_val) > 4:
        print(f"The number should be in the range: {range_}") 

#If the length is 4, 3, 2 or 1, proceed
elif len(split_val) <= 4:
    
    # Ones: print if the number inputted is in the range [-9, 9]:
    if len(split_val) == 1:
        minus += " " + num_words[ones]
        
    # Tens: print if the number inputted is in the range [-99, 99]:
    if len(split_val) == 2:
        if split_val[0] == 1:
            minus += " " + Suffix_num[ones]
        elif split_val[1] == 0:
            minus += " " + num_with_0_words[tens]
        
        else:
            minus += " " + num_with_0_words[tens] +" "+ num_words[ones]

    # Hundreds: print if the number is in the range [-999, 999]:
    if (len(split_val) == 3):
        if split_val[1] == 1:
            minus += " " + num_words[hundred] + " thousand " + Suffix_num[split_val[2]]
        elif split_val[2] == 0:
            minus += " " + num_words[hundred] + " hundreds " + num_with_0_words[tens]
            
        else:
            minus += " " + num_words[hundred] + " hundreds " + num_with_0_words[tens] +" "+ num_words[ones]

    # Thousands: print if the number is in range [-9999, 9999]:
    if (len(split_val) == 4):
        if (split_val[1] == 0) and (split_val[2] == 1):
            minus += " " + num_words[thousands] +" thousand " + num_words[hundred] +""+ Suffix_num[ones]
        elif (split_val[1] == 0) and (split_val[3] == 0):
            minus += " " + num_words[thousands] +" thousand "+ num_with_0_words[tens] 
        elif split_val[3] == 0:
            minus += " " + num_words[thousands] +" thousand "+ num_words[hundred] +" hundred and "+ num_with_0_words[tens]
        elif split_val[2] == 1:
            minus += " " + num_words[thousands] +" thousand "+ num_words[hundred] +" hundred and "+ Suffix_num[split_val[3]] 
            
        elif split_val[2:3] == 0:
            minus += " " + num_words[thousands] +" thousand "+ num_words[hundred] 
            
            
        elif split_val[1:3] == [0,0]:
            minus += " " + num_words[thousands]+""+" thousand "+""+ num_words[ones] 
        elif split_val[2] == 0:
            minus += " " + num_words[thousands] +" thousand "+ num_words[hundred] +" hundred and "+ num_words[ones] 
        else:
            minus += " " + num_words[thousands] +" thousand "+ num_words[hundred] +" hundred and "+ num_with_0_words[tens] +"-"+ num_words[ones]
    
    print(minus) # if the string minus is stored will be connected to output under the satisfied condition:

    

      
```
